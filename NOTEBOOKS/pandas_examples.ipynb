```python
cols = ['alpha', 'beta', 'gamma', 'delta', 'epsilon']
index = ['a', 'b', 'c', 'd', 'e', 'f']

values = [
    [100, 110, 120, 130, 140],
    [200, 210, 220, 230, 240],
    [300, 310, 320, 330, 340],
    [400, 410, 420, 430, 440],
    [500, 510, 520, 530, 540],
    [600, 610, 620, 630, 640],
]


df = pd.DataFrame(values, index=index, columns=cols)
df
```

```python
def times_ten(x):
    return x * 10

df['zeta'] = df['delta'] * df['epsilon'] # product of two columns
df['eta'] = times_ten(df.alpha) # user-defined function
df['theta'] = df.sum(axis=1)  # sum each row
df['iota'] = df.mean(axis=1)  # avg of each row
df['kappa'] = df.loc[:,'alpha':'gamma'].mean(axis=1)
# column kappa is avg of selected columns
df
```

```python
df.drop(['beta', 'zeta'], axis=1, inplace=True)
df
```

```python
df.drop(['b', 'c','d'], inplace=True)
df
```
